```python
import pandas as pd
path ="/content/drive/MyDrive/DATA_MINING_DATASETS/playlist_2010to2022.csv"
df = pd.read_csv(path)
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

missing_values
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Get descriptive statistics for the numerical columns
descriptive_stats = df_cleaned.describe()

descriptive_stats
```

```python
grouped_data = df_cleaned.groupby('year').track_popularity.mean().reset_index()
```

```python

# Aggregate the data to get the average track popularity for each year
popularity_by_year = df_cleaned.groupby('year')['track_popularity'].mean().reset_index()

popularity_by_year
```

```python
from IPython.core.display import display, HTML

#@title Convert the aggregated data to a format suitable for JavaScript
data_js = [{"year": str(row["year"]), "popularity": row["track_popularity"]} for _, row in popularity_by_year.iterrows()]

# D3.js code to create a bar chart
d3_script = f"""
<script src="https://d3js.org/d3.v6.min.js"></script>
<div id="d3-container"></div>
<script>
    // Set the dimensions and margins of the graph
    var margin = {{top: 10, right: 30, bottom: 40, left: 50}},
        width = 600 - margin.left - margin.right,
        height = 400 - margin.top - margin.bottom;

    // Append the SVG object to the body of the page
    var svg = d3.select("#d3-container")
      .append("svg")
        .attr("width", width + margin.left + margin.right)
        .attr("height", height + margin.top + margin.bottom)
      .append("g")
        .attr("transform", "translate(" + margin.left + "," + margin.top + ")");

    // Data
    var data = {str(data_js)};

    // X axis
    var x = d3.scaleBand()
      .range([ 0, width ])
      .domain(data.map(function(d) {{ return d.year; }}))
      .padding(0.2);
    svg.append("g")
      .attr("transform", "translate(0," + height + ")")
      .call(d3.axisBottom(x))

    // Y axis
    var y = d3.scaleLinear()
      .domain([0, 100])
      .range([ height, 0]);
    svg.append("g")
      .call(d3.axisLeft(y));

    // Bars
    svg.selectAll("mybar")
      .data(data)
      .enter()
      .append("rect")
        .attr("x", function(d) {{ return x(d.year); }})
        .attr("y", function(d) {{ return y(d.popularity); }})
        .attr("width", x.bandwidth())
        .attr("height", function(d) {{ return height - y(d.popularity); }})
        .attr("fill", "#69b3a2")
</script>
"""

# Display the D3.js bar chart in Colab
display(HTML(d3_script))
```

```python
# @title Convert the danceability, energy, and valence data to a format suitable for JavaScript
danceability_data_js = [{"value": row["danceability"]} for _, row in df_cleaned.iterrows()]
energy_data_js = [{"value": row["energy"]} for _, row in df_cleaned.iterrows()]
valence_data_js = [{"value": row["valence"]} for _, row in df_cleaned.iterrows()]

# D3.js code to create density plots for danceability, energy, and valence
d3_density_script = f"""
<script src="https://d3js.org/d3.v6.min.js"></script>
<div id="d3-danceability-container" style="float:left; width:33%;"></div>
<div id="d3-energy-container" style="float:left; width:33%;"></div>
<div id="d3-valence-container" style="float:left; width:33%;"></div>
<script>
    // Function to draw density plot
    function drawDensityPlot(container, data, color, title) {{
        var margin = {{top: 10, right: 30, bottom: 30, left: 50}},
            width = 200 - margin.left - margin.right,
            height = 400 - margin.top - margin.bottom;

        // Create SVG
        var svg = d3.select(container)
          .append("svg")
            .attr("width", width + margin.left + margin.right)
            .attr("height", height + margin.top + margin.bottom)
          .append("g")
            .attr("transform", "translate(" + margin.left + "," + margin.top + ")");

        // X axis
        var x = d3.scaleLinear()
            .domain([0, 1])
            .range([0, width]);
        svg.append("g")
            .attr("transform", "translate(0," + height + ")")
            .call(d3.axisBottom(x));

        // Y axis
        var y = d3.scaleLinear()
            .range([height, 0])
            .domain([0, 5]);
        svg.append("g")
            .call(d3.axisLeft(y));

        // Compute kernel density estimation
        var kde = kernelDensityEstimator(kernelEpanechnikov(0.05), x.ticks(40))
        var density =  kde(data.map(function(d){{ return d.value; }}))

        // Plot the area
        svg.append("path")
            .attr("class", "mypath")
            .datum(density)
            .attr("fill", color)
            .attr("opacity", ".8")
            .attr("stroke", "#000")
            .attr("stroke-width", 1)
            .attr("d",  d3.line()
                .curve(d3.curveBasis)
                .x(function(d) {{ return x(d[0]); }})
                .y(function(d) {{ return y(d[1]); }})
            );

        // Add title
        svg.append("text")
            .attr("x", (width / 2))
            .attr("y", 0 - (margin.top / 2))
            .attr("text-anchor", "middle")
            .style("font-size", "12px")
            .style("font-weight", "bold")
            .text(title);
    }}

    // Compute kernel density estimation
    function kernelDensityEstimator(kernel, X) {{
        return function(V) {{
            return X.map(function(x) {{
                return [x, d3.mean(V, function(v) {{ return kernel(x - v); }})];
            }});
        }};
    }}

    function kernelEpanechnikov(k) {{
        return function(v) {{
            return Math.abs(v /= k) <= 1 ? 0.75 * (1 - v * v) / k : 0;
        }};
    }}

    // Data
    var danceabilityData = {str(danceability_data_js)};
    var energyData = {str(energy_data_js)};
    var valenceData = {str(valence_data_js)};

    // Draw the density plots
    drawDensityPlot("#d3-danceability-container", danceabilityData, "#69b3a2", "Danceability");
    drawDensityPlot("#d3-energy-container", energyData, "#ff9999", "Energy");
    drawDensityPlot("#d3-valence-container", valenceData, "#66c2a5", "Valence");
</script>
"""

# Display the D3.js density plots in Colab
display(HTML(d3_density_script))
```
