```python
import pandas as pd
path = "/content/drive/MyDrive/DATA_MINING_DATASETS/youtubers_df.csv"
youtubers_df = pd.read_csv(path)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set_style("whitegrid")

# Plotting distribution of subscribers for top YouTubers
plt.figure(figsize=(12, 6))
sns.histplot(youtubers_df['Suscribers'], bins=30, color='skyblue', kde=True)
plt.title('Distribution of Subscribers for Top YouTubers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Number of YouTubers')
plt.show()
```

```python
# @title Grouping by Country and summing the subscribers
country_subs = youtubers_df.groupby('Country')['Suscribers'].sum().sort_values(ascending=False).head(10)

# Plotting the top countries by total subscribers
plt.figure(figsize=(12, 6))
sns.barplot(y=country_subs.index, x=country_subs.values, palette='viridis')
plt.title('Top 10 Countries by Total Number of Subscribers')
plt.xlabel('Total Number of Subscribers')
plt.ylabel('Country')
plt.show()
```

```python
# @title Plotting distribution of likes and comments for top YouTubers
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Likes distribution
sns.histplot(youtubers_df['Likes'], bins=30, color='salmon', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Likes for Top YouTubers')
axes[0].set_xlabel('Number of Likes')
axes[0].set_ylabel('Number of YouTubers')

# Comments distribution
sns.histplot(youtubers_df['Comments'], bins=30, color='lightgreen', kde=True, ax=axes[1])
axes[1].set_title('Distribution of Comments for Top YouTubers')
axes[1].set_xlabel('Number of Comments')
axes[1].set_ylabel('Number of YouTubers')

plt.tight_layout()
plt.show()
```

```python
# @title Compute the correlation matrix
corr = youtubers_df[['Suscribers', 'Likes', 'Comments', 'Visits']].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
